```python
from bs4 import BeautifulSoup
import requests
from datetime import datetime, timedelta

articles = []

start_date = datetime(2023, 2, 1)
end_date = datetime(2023, 3, 16)

while start_date <= end_date:
    date = start_date.strftime('%Y%m%d')
    page_no = 1  # start with page 1
    while True:  # iterate until there are no more pages
        URL = f"https://news.daum.net/breakingnews/economic?page={page_no}&regDate={date}"
        res = requests.get(URL)
        soup = BeautifulSoup(res.text, 'html.parser')

        article_links = soup.select('.list_new2 a.link_txt, .list_allnews a.link_txt')
        if not article_links:  # no article links found, exit the loop
            break

        for link in article_links:
            article_url = link.get('href')
            if 'http' not in article_url:
                continue

            article_response = requests.get(article_url)
            article_soup = BeautifulSoup(article_response.text, 'html.parser')
            article_title = article_soup.select('h3.tit_view')[0].text
            article_reporter = article_soup.select('span.txt_info')[0].text
            article_date = article_soup.select('span.num_date')[0].text
            article_text = article_soup.select('div.article_view')[0].text

            articles.append(article_title.strip().replace('\n', ' '))
            articles.append(article_reporter.strip().replace('\n', ' '))
            articles.append(article_date.strip().replace('\n', ' '))
            articles.append(article_text.strip().replace('\n', ' '))

        page_no += 1  # move to the next page

    start_date += timedelta(days=1)
```
